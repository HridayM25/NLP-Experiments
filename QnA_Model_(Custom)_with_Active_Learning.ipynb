```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
```

```python
filename=input("Enter the name of the csv file: ")
```

```python
df=pd.read_csv(filename)
```

```python
df.rename(columns={"Selection Procedure (1st round, 2nd round, Complete Description)":"Procedure"}, inplace=True)
```

```python
p=df.Procedure
```

```python
p=p.dropna()
```

```python
def clean_for_test(context):
  context= BeautifulSoup(context, "lxml").text
  context=context.replace("\n", "")
  context=context.replace("\t", "")
  context=context.replace("\r", "")
  return context
```

```python
def rounds(context,question):
  context=clean_for_test(context)
  return new_reader.predict_on_texts(question, [context])
```

Code for performing Active Learning on this dataset.

```python
manual_annotation=[]
training_new=[]
answers_json=[]
```

```python
def active_learner(context,question):
    answer=rounds(context,question)
    score=answer['answers'][0].score
    if score<0.65:
      return 1
    else:
      training_new.append(context)
      answers_json.append(answer)
      return 0
```

This will add the entry number corresponding to GYAN entries to the list of manual_annotation

```python
def pipeline (start,end,question):
  for entry in range (start, end):
    context=p.iloc[entry]
    print(entry)
    flag=active_learner(context, question)
    if flag==1:
      manual_annotation.append({question:entry})
```
