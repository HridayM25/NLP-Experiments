```python
from haystack.nodes import FARMReader
from haystack.utils import fetch_archive_from_http
```

```python
#/content/data/answers.json is the correct path
```

```python
reader = FARMReader(model_name_or_path="distilbert-base-uncased-distilled-squad", use_gpu=True)
data_dir = "data/"
# data_dir = "PATH/TO_YOUR/TRAIN_DATA"
reader.train(data_dir=data_dir, train_filename="answers.json", use_gpu=True, n_epochs=15, save_dir="my_model")
```

```python
#save the model as my_model. Can use it later for more gyan reviews.
reader.save(directory="my_model")
```
